# Mellea Quickstart

Mellea is a library for writing generative programs.

Mellea takes us back to the future of computing by formally introducing the concept of generative programs - software systems that strategically integrate calls to Large Language Models (LLMs) - and the demanding engineering required to make them reliable. The fundamental challenge we address is how to safely and predictably harness the powerful but inherently stochastic operations of LLMs within traditionally deterministic codebases.

In this recipe, we will introduce use the [Mellea](https://mellea.ai) library and write a introductory generative program.

Mellea supports several different LLM runtimes, called `backends`, to access an LLM. In this recipe, we will use the [LiteLLM](https://docs.litellm.ai/docs/) [backend](https://github.com/generative-computing/mellea/blob/main/mellea/backends/litellm.py). We can use the LiteLLM backend with providers supported by LiteLLM such as IBM [watsonx.ai](https://www.ibm.com/products/watsonx-ai). In this recipe, you will use the IBM® [Granite](https://www.ibm.com/granite) model now available on watsonx.ai.

Now we can create a MelleaSession using the LiteLLM backend connected to the Granite model on watsonx.ai.

```python
from mellea import start_session, MelleaSession

m: MelleaSession = start_session(
        backend_name="litellm",
        model_id="watsonx/ibm/granite-4-h-small",
    )
```

## Step 4. Write an email with a simple instruction to the model

In this step, we request the model to write an email using one function call.

```python
email = m.instruct("Write an email inviting the interns to a lunch party.")

print(email)
```

## Step 5. Write an email using user variables

In this step, we write a function which takes input variables and we pass those variables to Mellea along with a template instruction to the model. Mellea will format the template using the variables into the instruction for the model.

```python
def write_email(m: MelleaSession, name: str, notes: str) -> str:
    email = m.instruct(
        "Write an email to {{name}} using the notes following: {{notes}}.",
        user_variables={"name": name, "notes": notes},
    )
    return str(email)

email = write_email(
        m,
        "Olivia",
        "Olivia helped the lab over the last few weeks by organizing intern events, advertising the speaker series, and handling issues with snack delivery.",
    )

print(email)
```
